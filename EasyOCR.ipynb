```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 8,16
```

```python
reader = easyocr.Reader(['en'])
```

```python
Image('img.jpg')
```

```python
output = reader.readtext('img.jpg')
```

```python
output
```

```python
output[0][1]
```

```python
output[1][1]
```

```python
cord = output[-1][0]
```

```python
cord1 = output[-2][0]
```

```python
x_min1,y_min1 = [min(idx) for idx in zip(*cord1)]
x_max1,y_max1 = [max(idx) for idx in zip(*cord1)]
```

```python
x_min,y_min = [min(idx) for idx in zip(*cord)]
```

```python
x_max,y_max = [max(idx) for idx in zip(*cord)]
```

```python
image = cv2.imread('img.jpg')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(255,0,0),2)
cv2.rectangle(image,(x_min1,y_min1),(x_max1,y_max1),(255,0,0),2)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```
