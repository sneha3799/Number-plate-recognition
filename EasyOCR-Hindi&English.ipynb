```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
```

```python
rcParams['figure.figsize'] = 8,16
```

```python
reader = easyocr.Reader(['en','hi'])
```

```python
file_name = "pic.jpg"
```

```python
Image(file_name)
```

```python
output = reader.readtext(file_name)
```

```python
output
```

```python
cord = output[2][0]
cord1 = output[0][0]
cord2 = output[1][0]
cord3 = output[3][0]
```

```python
x_min, y_min = [int(min(idx)) for idx in zip(*cord)]
x_max,y_max = [int(max(idx)) for idx in zip(*cord)]

x_min1, y_min1 = [int(min(idx)) for idx in zip(*cord1)]
x_max1,y_max1 = [int(max(idx)) for idx in zip(*cord1)]

x_min2, y_min2 = [int(min(idx)) for idx in zip(*cord2)]
x_max2,y_max2 = [int(max(idx)) for idx in zip(*cord2)]

x_min3, y_min3 = [int(min(idx)) for idx in zip(*cord3)]
x_max3,y_max3 = [int(max(idx)) for idx in zip(*cord3)]
```

```python
image = cv2.imread(file_name)
```

```python
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(255,0,0),2)
cv2.rectangle(image,(x_min1,y_min1),(x_max1,y_max1),(255,0,0),2)
cv2.rectangle(image,(x_min2,y_min2),(x_max2,y_max2),(255,0,0),2)
cv2.rectangle(image,(x_min3,y_min3),(x_max3,y_max3),(255,0,0),2)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```
